```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,9)
y = np.arange(4,12)

plt.subplot(2,2,1)
plt.plot(x , x*x , "blue")

plt.subplot(2,2,2)
plt.plot(x , np.sqrt(x) , "red")

plt.subplot(2,2,3)
plt.plot(x , np.exp(x) , "yellow")

plt.subplot(2,2,4)
plt.plot(x , np.log10(x) , "black")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,9)
y = np.arange(4,12)

fig, ax = plt.subplots(2,2)
ax[0,0].plot(x , np.sqrt(x) , "red")
ax[0,1].plot(x , x*x , "blue")
ax[1,0].plot(x , np.log10(x) , "black")
ax[1,1].plot(x , np.exp(x) , "green")
plt.show()
```
